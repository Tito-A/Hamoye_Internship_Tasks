# Multivariate prediction

```python
# selecting the columns

df['ds'] = pd.to_datetime(df['Time'])
df_2 = df[['ds', 'global_active_power', 'Global_reactive_power', 'Voltage',
         'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']]
```

```python
x = len(df)
test_size = 300

train_df = df_2.head(x - test_size)
test_df = df_2.tail(test_size)
```

```python
import prophet

model = prophet.Prophet()
model.add_regressor('Global_reactive_power')
model.add_regressor('Voltage')
model.add_regressor('Global_intensity')
model.add_regressor('Sub_metering_1')
model.add_regressor('Sub_metering_2')
model.add_regressor('Sub_metering_3')

model.fit(train_df)
```

```python
future = model.make_future_dataframe(periods=test_size)
future['Global_reactive_power'] = df['Global_reactive_power'].values
future['Voltage'] = df['Voltage'].values
future['Global_intensity'] = df['Global_intensity'].values
future['Sub_metering_1'] = df['Sub_metering_1'].values
future['Sub_metering_2'] = df['Sub_metering_2'].values
future['Sub_metering_3'] = df['Sub_metering_3'].values

forecast = model.predict(future)
```

```python
fig = model.plot(forecast)
```
